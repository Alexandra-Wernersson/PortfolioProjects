```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None
```

```python
# Now we need to read in the data
df = pd.read_csv(r'/Users/alexandrawernersson/Downloads/movies.csv')
```

```python
# Look at data

df.head()
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
# Look at correaltions; Hypothesis: budget will have high correlation, company high correlation
```

```python
# Scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross earnings')
plt.xlabel('Budget [Dollars]')
plt.ylabel('Gross [Dollars]')
plt.show()
```

```python
df.head()
```

```python
# Plot the budget vs gross

sns.regplot(x='budget', y='gross', data = df, scatter_kws={"color":"red" }, line_kws = {"color":"blue"})
```

```python
# Let's start looking at correlations
numeric_df = df.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_df.corr(method='pearson') #Can use Pearosn, Kendall, Spearman
```

```python
correlation_matrix
```

```python
# High correlation between budget and gross
```

```python
plt.title('Correlation matrix for numeric features')
plt.xlabel('Movie features')
plt.ylabel('Movie features')
sns.heatmap(correlation_matrix, annot=True)
```

```python
df_numerized = df
for col_name in df_numerized:
    if df_numerized[col_name].dtype=="object":
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
```

```python
df_numerized
```

```python
df
```

```python
correlation_matrix = df_numerized.corr()
plt.title('Correlation matrix for numeric features')
plt.xlabel('Movie features')
plt.ylabel('Movie features')
sns.heatmap(correlation_matrix, annot=True)
```

```python
corr_pairs = correlation_matrix.unstack()
```

```python
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

print(high_corr)
```

```python
# Votes and budget have the highest correlation to gross earnings
# Company has low correlation
```
